```python
import matplotlib.pyplot as plt
import numpy as np

# создаём данные на основе которых будем строить графики
np.random.seed(19680801)
first_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

np.random.seed(27930017)
second_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x4', 'x5', 'x6']

# задаём два соседствующих графика
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# стоим сам boxplot
bplot1 = axes[0].boxplot(first_data,
                         vert=True,
                         patch_artist=True,
                         labels=labels)

bplot2 = axes[1].boxplot(second_data,
                         vert=True,
                         patch_artist=True,  
                         labels=labels)  

# раскрашиваем
colors = ['white', 'gray', 'black']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
    
# добавляем подписи к осям
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xlabel('Samples')
    ax.set_ylabel('Observed values')
   
Samples = ['1','2','3']

plt.legend(Samples,loc='upper left')

plt.show()
```
